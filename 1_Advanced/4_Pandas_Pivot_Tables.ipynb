```python
# Importing Libraries
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleaning 
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

# Count of Job Titles

```python
df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
df.groupby('job_title_short').size()
```

```python
df.pivot_table(
    values='salary_year_avg', 
    columns='job_title_short', 
    index = 'job_country', 
    aggfunc='median')
```

```python
df.groupby('job_title_short')[['salary_year_avg']].median()
```

```python
#.value_counts() â†’ counts how many times each unique country appears in that column, and sorts the result in descending order (most frequent first).

#.head(6) â†’ takes only the top 6 countries from that list (the 6 with the highest counts).

#.index â†’ returns just the names (the country names) of those top 6, not their counts.

top_countries = df['job_country'].value_counts().head(6).index

```

How do median data-related salaries (Data Analyst, Data Engineer, and Data Scientist) compare across the top six countries with the most job postings?

```python
top_countries = df['job_country'].value_counts().head(6).index

df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar')

plt.ylabel('Median salary ($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
